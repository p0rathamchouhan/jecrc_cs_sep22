```python
import matplotlib.pyplot as plt
import numpy as np
```

# Type of Graphs
1. plot -           line graph, between two prameters
2. bar -            bar graph of single parameter (optional catagorical data)
3. hist -           histogram of a bunch of data i.e. frequency of single parameter
4. scatter -        dot graph between two paramteters
5. boxplot / box -  statistical overview in form of bax for a single parameter
                    mean,outliers, qunatiles(25%,75%), 10R
                    (optional - categorical data)
6. stem -           more like bar graph , Lines instead of bars
7. violinplot -     violin shaped statistical overview graph
                    spread of data (more like hist)
                    single parameter (optional cat-data)

```python
x= np.linspace(1,5,11)
x1=np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y=x**3
```

```python
y
```

```python
plt.plot(x,y)
```

```python
y2=x**4 
plt.plot(x,y2)
plt.show()
```

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()
```

```python
plt.hist(x)
plt.title('Histogram')
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter Graph')
plt.grid()
plt.xlabel('x -- >')
plt.ylabel('y -- >')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()

plt.boxplot(x)
plt.title('Box Plot')
plt.show()
```

```python
plt.stem(x,y)
plt.title('Stem Graph')
plt.show()

plt.violinplot(y)
plt.show()
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig =plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markersize=15,markerfacecolor='purple')
plt.plot(x,y2,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
