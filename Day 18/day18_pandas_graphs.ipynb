# Graph Function
1. plot               - Simple line plot
2. plot.area          - filled version of line plot
3. plot.bar/plot.barh - bar graph
4. plot.box           - box plot
5. plot.density       - Density line (smooth curved)
6. plot.hexbin        - hexagon version of scatter with desnity
7. plot.hist          -  histogram
8. plot.kde           - kernel denstiy estimator
9. plot.pie           - pie chart of categorical data
10. plot.scatter      - scatter graph

```python
df2.plot();plt.title('Line Plot'); plt.show()
df2.plot.area();plt.title('Area Plot'); plt.show()
df2.plot.bar();plt.title('Bar Plot'); plt.show()
df2.plot.box();plt.title('Box Plot'); plt.show()
df2.plot.barh();plt.title('Bar horizontal Plot'); plt.show()
df2.plot.density();plt.title('DEnsity Plot'); plt.show()
df2.plot.kde();plt.title('KDE Plot'); plt.show()
df2.plot.hist();plt.title('HIstohram Plot'); plt.show()
df2.plot.scatter(x='a', y='b');plt.title( 'Scatter Plot'); plt.show()
df2.plot.hexbin(x='a',y='b',C='c',gridsize=6);plt.title('Hexbin PLot')
```

```python
import seaborn as sns
```

#### Categories of function
# 1. Categorical Plots  (hue can be used | comparison with categorical data)
    a. countplot   (single parameter)
    b. barplot      (onr parametre should be categorical and one should be numerical )
    c. boxplot
    d. violinplot
    e. swarmplot
# 2. Distribution plots
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
# 3. Matrix Plots
    a. heatmap
    b. clusterplot
    c. pairplot  (same as above one)
# 4. Regression Plots
    a. lmplot
    b. regplot
# 5. Grid Plots
    a. PairGrid
    b. JointGrid
    c. FaceGrid

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.mean)
```

```python
sns.set_style('whitegrid')
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.set_style('whitegrid')
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```
