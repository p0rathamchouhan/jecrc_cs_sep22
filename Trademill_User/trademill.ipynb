# Bivariate Analysis

1. boxplot -cat,num
2. jointplot-num,num
3. scatterplot-num,num   (x)
4. violinplot-cat,num
5. lineplot-num,num
6. kdeplot/rudplot-num,num
7. regplot-num,num       (x)
8. barplot-cat,num
9. swarmplot-cat,num     (X)
10. pointplot-num,num
11. rugplot-num,num      (x)

```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.sublot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.show()
```

```python
sns.jointplot(data=data,x=col,y=ncol)
    plt.subplot(1,2,2)
    sns.jointplot(data=data,x=col,y=ncol)
```

```python
for col in num_cols:
    for ncol in num_cols:
        if(col==ncol):
            continue
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    
    sns.pairplot(data=data[num_cols])
```

```python
pg=sns.PairGrid(data=data[num_cols])
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
```

```python
for col in cat_cols:
    if col=='Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.show()
```
